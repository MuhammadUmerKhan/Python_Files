```python
import pandas as pd
```

```python
data = pd.read_csv('./canada_per_capita_income.csv')
data.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
X = data[['year']]
y = data[['per capita income (US$)']]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=X, y=y, marker='+', color='red')
```

```python
lr.fit(X, y)
```

```python
lr.predict([[2020]])
```

```python
plt.scatter(x=X, y=y, marker='+', color='red')
plt.plot(X, lr.predict(X), color='green', linewidth=2, label='Best fit line')
```
