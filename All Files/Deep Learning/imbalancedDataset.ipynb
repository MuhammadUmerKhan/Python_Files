```python
import numpy as np
def ANN(x_train, y_train, x_test, y_test, loss, weight):
    model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26, ), activation = 'relu'), # Input Layer
    keras.layers.Dense(15, activation = 'relu'), # Hidden Layer
    keras.layers.Dense(1, activation = 'sigmoid')  # Output Layer
    ])


    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])

    if weight == -1:
        model.fit(x_train, y_train, epochs=100)
    else:
        model.fit(x_train, y_train, epochs=100, class_weight=weight)


    print(model.evaluate(x_test, y_test))
    y_pred = model.predict(x_test)
    y_pred = np.round(y_pred)
    print(classification_report(y_test, y_pred))
    return y_pred
```

## Mitigating Skewdness of Data

#### Method 1: Undersampling

```python
count_class_0, count_class_1 = df1['Churn'].value_counts()
```

```python
df_class_0 = df2[df2['Churn'] == 0]
df_class_1 = df2[df2['Churn'] == 1]
```

```python
count_class_0, count_class_1
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)
```

```python
print(df_test_under['Churn'].value_counts())
```

```python
X = df_test_under.drop(columns='Churn')
Y = df_test_under['Churn']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=15, stratify=Y)
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(x_train, y_train, x_test, y_test, 'binary_crossentropy', -1)
```

```python
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_test_over = pd.concat([df_class_0, df_class_1_over], axis=0)
df_test_over['Churn'].value_counts()
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

#### Method_3: SMOTE

```python
from imblearn.over_sampling import SMOTE
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
y_train.value_counts()
```

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
y_train.value_counts()
```

```python
df3 = X_train.copy()
df3['Churn'] = y_train
```

```python
df3.head()
```

```python
df3_class0 = df3[df3.Churn==0]
df3_class1 = df3[df3.Churn==1]
```

```python
def get_train_batch(df_majority, df_minority, start, end):
    df_train = pd.concat([df_majority[start:end], df_minority], axis=0)

    X_train = df_train.drop('Churn', axis='columns')
    y_train = df_train.Churn
    return X_train, y_train    
```

```python
X_train, y_train = get_train_batch(df3_class0, df3_class1, 0, 1495)

y_pred1 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
len(y_pred1)
```

```python
y_pred_final = y_pred1.copy()
for i in range(len(y_pred1)):
    n_ones = y_pred1[i] + y_pred2[i] + y_pred3[i]
    if n_ones>1:
        y_pred_final[i] = 1
    else:
        y_pred_final[i] = 0
```

#### Method_4 Use of Ensemble with undersampling

```python
# Regain Original features and labels
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
df3 = X_train.copy()
df3['Churn'] = y_train
df3_class0 = df3[df3.Churn==0]
df3_class1 = df3[df3.Churn==1]
def get_train_batch(df_majority, df_minority, start, end):
    df_train = pd.concat([df_majority[start:end], df_minority], axis=0)

    X_train = df_train.drop('Churn', axis='columns')
    y_train = df_train.Churn
    return X_train, y_train    

X_train, y_train = get_train_batch(df3_class0, df3_class1, 0, 1495)
```

```python
model_log.fit(X_train, y_train)
log_pred = model_log.predict(X_test)
log_pred1 = np.round(log_pred)
print(classification_report(y_test, log_pred1))
```
