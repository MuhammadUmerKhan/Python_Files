```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
data = pd.read_csv("./insurance_data.csv")
data.head()
```

```python
X_train, X_test, y_train, y_tes = train_test_split(data[['age', 'affordibility']], data['bought_insurance'], test_size=0.2, random_state=42)
```

```python
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled['age'] = X_train_scaled['age'] / 100
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

##### Model Building: First build a model in keras/tensorflow and see what weights and bias values it comes up with. We will than try to reproduce same weights and bias in our plain python implementation of gradient descent. Below is the architecture of our simple neural network

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.BinaryCrossentropy, metrics=['accuracy'])
```

```python
model.fit(X_train_scaled, y_train, epochs=5000)
```

```python
model.evaluate(X_test_scaled,y_test)
```
